```python
import numpy as np
def confusion_metrix(y_true,y_pred,lables):
    cm=np.zeros((len(lables),len(lables)),dtype=int)
    for t,p in zip(y_true,y_pred):
        cm[lables[t],lables[p]]+=1
    return cm    
y_true=[0, 1, 0, 1, 0, 1, 1, 0]      
y_pred=[0, 0, 0, 1, 0, 1, 1, 1] 
lables=[0,1]
cm=confusion_metrix(y_true,y_pred,lables)
print(cm)
```

```python
tp =0
tn =0
fp=0
fn=0
for i,j in zip(y_true,y_pred):
    if i == 0 and j == 0:
        tn += 1
    elif i == 0 and j == 1:
        fp += 1
    elif i == 1 and j == 0:
        fn += 1
    else:
        tp += 1
print(tn,fp,fn,tp)
```

```python
import matplotlib.pyplot as plt


x1 = np.random.randint(40, 95, size=(55,10), dtype=int)
x2 = np.random.randint(5, 95, size= (45,10), dtype=int)
x_train = np.concatenate((x1,x2))

```
