```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define function and gradient
def f(x, y):
    return x**3 + y**3 + x**2*y + y**2*x
```

```python
def grad_f(x, y):
    df_dx = 3*x**2 + 2*x*y + y**2
    df_dy = 3*y**2 + 2*y*x + x**2
    return np.array([df_dx, df_dy])
```

```python

# Steepest Descent Algorithm
alpha = 0.01  # Learning rate
iterations = 100
x_vals, y_vals = [], []
```

```python
# Initial guess
x, y = 2.0, -2.0 
```

```python
for i in range(iterations):
    grad = grad_f(x, y)
    x, y = x - alpha * grad[0], y - alpha * grad[1]
    x_vals.append(x)
    y_vals.append(y)
```

```python
# Compute function values for convergence plot
f_vals = [f(x_vals[i], y_vals[i]) for i in range(iterations)]
```

```python
# Plot the convergence
plt.figure(figsize=(8, 5))
plt.plot(range(iterations), f_vals, marker='o', linestyle='-', color='b')
plt.xlabel("Iterations")
plt.ylabel("Function Value")
plt.title("Convergence of Steepest Descent Gradient Algorithm")
plt.grid()
plt.show()
```
