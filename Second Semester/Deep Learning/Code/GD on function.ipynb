```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and its gradient
def f(x, y):
    return x**3 + y**3 + x**2 * y + y**2 * x
```

```python
def grad_f(x, y):
    df_dx = 3*x**2 + 2*x*y + y**2
    df_dy = 3*y**2 + 2*y*x + x**2
    return np.array([df_dx, df_dy])
```

```python

# Gradient Descent Algorithm
def gradient_descent(learning_rate=0.01, max_iters=100, tol=1e-6):
    x, y = np.random.uniform(-2, 2, 2)  # Random initial point
    history = [(x, y, f(x, y))]  # Store points for plotting
    
    for i in range(max_iters):
        grad = grad_f(x, y)
        x_new = x - learning_rate * grad[0]
        y_new = y - learning_rate * grad[1]
        
        if np.linalg.norm([x_new - x, y_new - y]) < tol:
            break  # Convergence check
        
        x, y = x_new, y_new
        history.append((x, y, f(x, y)))
    
    return np.array(history)
```

```python
# Run gradient descent
learning_rate = 0.01
history = gradient_descent(learning_rate)
```

```python
# Plot convergence
plt.figure(figsize=(8, 5))
plt.plot(history[:, 2], label='Function Value')
plt.xlabel('Iteration')
plt.ylabel('f(x, y)')
plt.title('Convergence of Gradient Descent')
plt.legend()
plt.grid()
plt.show()
```
